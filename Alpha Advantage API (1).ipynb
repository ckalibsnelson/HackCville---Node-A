```python
#Download CSV file from Alpha Advantage API: https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=MSFT&interval=5min&apikey=demo&datatype=csv
#Microsoft (MSFT) stock data per Alpha Advantage
#Connect to API via CSV file
import csv
import requests

CSV_URL = ('https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=MSFT&interval=5min&apikey=' + API_KEY + '&datatype=csv')

with requests.session() as s:
    download = s.get(CSV_URL)
    
    decoded_content = download.content.decode('utf-8')
    
    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    for row in my_list:
        print(row)
```

```python
#Create dataframe out of list from CSV file
import pandas as pd
MSFT_CSV = pd.DataFrame(my_list[1:],columns = my_list[0])
MSFT_CSV.head()
```

```python
#Using a Stock Market API: https://medium.com/@Keithweaver_/using-a-stock-market-api-4ce65b7c67ac
#Connect to API via JSON file
r = requests.get('https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=MSFT&apikey=' + API_KEY)
if (r.status_code == 200):
    MSFT = r.json()
    MSFT_All_Days = MSFT['Time Series (Daily)']
    MSFT_Single_Date = MSFT_All_Days['2019-03-15']
    MSFT_open = float(MSFT_Single_Date['1. open'])
    MSFT_high = float(MSFT_Single_Date['2. high'])
    MSFT_low = float(MSFT_Single_Date['3. low'])
    MSFT_close = float(MSFT_Single_Date['4. close'])
    MSFT_volume = float(MSFT_Single_Date['5. volume'])
```
